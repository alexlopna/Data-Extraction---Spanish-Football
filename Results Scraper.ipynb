#Creamos una función para las temporadas disputadas
def Web_Scrapper(Web, clase):


 #Cargamos la web
    page= urllib.request.urlopen(Web)
    soup=BeautifulSoup(page, "lxml")
    right_table=soup.find('div', class_=clase)

#Contamos el número de columnas. Nos valdrá para automatizar el tamaño de la lista

    N_columnas=len(right_table.find_all("tr")[0].find_all("td")) #Número de columnas del dataset
    N_filas=len(right_table.find_all("tr")) #Numero de filas del dataset

#Generamos listas vacias (Equivalente a 17 columnas, se pueden añadir más si se quiere)   
    A=[]
    B=[]
    C=[]
    D=[]
    E=[]
    F=[]
    G=[]
    H=[]
    I=[]
    J=[]
    K=[]
    L=[]
    M=[]
    N=[]
    O=[]
    P=[]
    Q=[]

#Se crea una lista anidada vacía que contiene a su vez listas anidadas que corresponde a cada columna de la tabla
    lista=[A,B,C,D,E,F,G,H,I,J,K,L,M,N,O,P,Q] 
    
#Creamos nuestro listado vacío con las columnas de nuestra tabla seleccionada de la Web
    lista_columna=lista[0:N_columnas]
    
    for j in range(0,N_filas): #Recorremos todas las filas
        for i in range(0,N_columnas): #Recorremos las columnas
            lista_columna[i].append(right_table.find_all("tr")[j].find_all("td")[i].find(text=True)) #Va rellenando la lista
        #print(j,i)  


#Convertimos la lista en DataFrame
    df=pd.DataFrame(lista_columna[0],columns=["Equipo_Casa"]) # Creamos la primera columna
    df["Resultados"]=lista_columna[1]
    df["Equipo_Visitante"]=lista_columna[2]
    
    return df
